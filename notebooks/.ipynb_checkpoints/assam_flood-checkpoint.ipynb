```python
df=df.drop_duplicates()
df=df.reset_index(drop=True)
```

```python
# from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
```

```python
df.head()
```

```python

# Convert timestamp
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Check missing values
print(df.isnull().sum())

# Drop missing values
df.dropna(inplace=True)

print("After cleaning:", df.shape)
```

```python

df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['dayofweek'] = df['timestamp'].dt.dayofweek
```

```python

df['rainfall_x_soil'] = df['rainfall_12hr_sum'] * df['soil_moisture_percent']
df['water_vs_danger'] = df['river_water_level_m'] - df['danger_level_m']
df['rainfall_intensity'] = df['rainfall_12hr_sum'] / (df['rainfall_6hr_sum'] + 1)
```

```python

df = pd.get_dummies(df, columns=['district', 'river'], drop_first=True)
```

```python
# df.head(
df = df.sort_values('timestamp')
df.reset_index(drop=True, inplace=True)
```

```python
# df.info()
X = df.drop(['flood_next_24hr', 'timestamp'], axis=1)
y = df['flood_next_24hr']

print("Target distribution:")
print(y.value_counts())
```

```python


smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print("\nAfter SMOTE:")
print(y_resampled.value_counts())
```

```python

train_size = int(0.8 * len(X_resampled))

X_train = X_resampled.iloc[:train_size]
y_train = y_resampled.iloc[:train_size]

X_test = X_resampled.iloc[train_size:]
y_test = y_resampled.iloc[train_size:]
```

```python

model = XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)

model.fit(X_train, y_train)
```

# Model Accuracy

```python

from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
```

```python

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_prob))
```

```python

feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importances_
}).sort_values(by='importance', ascending=False)

feature_importance.head(10)
```
